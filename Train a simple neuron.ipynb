```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from IPython import display
import time
```

### Question 1

```python
# input arrays
x1 = np.array([0, 0, 1, 3, 2, 2])
x2 = np.array([0, 1, 0, 3, 2, 3])
# output array
t = np.array([0, 0, 0, 1, 1, 1])

# assigning random weights to the varaibles w1, w2 and w (bias)
w1 = np.random.uniform(-3.0, 3.0)
w2 = np.random.uniform(-3.0, 3.0)
w = np.random.uniform(-3.0, 3.0)

learning_rate = 0.5
max_iterations = 10000
k = np.linspace(-4, 4, num = 100)
lines = []

# creating a graph with blue circle and red square to differentiate class 1 and class 0 respectively
fig, ax = plt.subplots()
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.axis([-5.5, 5.5, -5.5, 5.5])
ax.scatter(x1[t == 1], x2[t == 1], c='blue', marker='o', label='Class 1')
ax.scatter(x1[t == 0], x2[t == 0], c='red', marker='s', label='Class 0')
ax.legend()

# function to calculate sigmoid
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

# loop iterating 10000 times to get the final classification line
for _ in range(max_iterations):
    
    # loop for calculating the error between the sigmoid output and target output
    error = 0
    for i in range(len(x1)):
        u = (w1 * x1[i]) + (w2 * x2[i]) + w
        o = sigmoid(u)
        error += (o - t[i])**2
    
    # if total error is less than or equal to 0.1 than break the loop
    if error <= 0.1:
        break
    
    # initializing varaibles for gradient descent of weights
    dw1 = 0
    dw2 = 0
    dw = 0
    
    # loop to find the derivation of the the input array w.r.t the weight
    for i in range(len(x1)):
        u = (w1 * x1[i]) + (w2 * x2[i]) + w
        o = sigmoid(u)
        dw1 += learning_rate * (x1[i] * (o - t[i]) * (o * (1 - o)))
        dw2 += learning_rate * (x2[i] * (o - t[i]) * (o * (1 - o)))
        dw += learning_rate * ((o - t[i]) * (o * (1 - o)))
        
    # updating weights
    w1 -= (1/5) * dw1
    w2 -= (1/5) * dw2
    w -= (1/5) * dw
    
    if dw1 == 0 or dw2 == 0 or dw == 0:
        break
        
    # updating the line using this equation
    if w2 != 0:
        l = ((-w1 * k) - w) / w2

    # plotting the updated line for every iteration
    line, = ax.plot(k, l, linestyle='--', c='g', alpha=0.2)
    lines.append(line)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.001)

# plotting the last line on the graph
ax.plot(k, l, c='b', linestyle='--', alpha=1)
plt.show()
```

```python
# input arrays
x1 = np.array([0, 0, 1, 3, 2, 2])
x2 = np.array([0, 1, 0, 3, 2, 3])
x3 = np.array([0, 0, 0, 3, 3, 2])
# output array
t = np.array([0, 0, 0, 1, 1, 1])

# assigning random weights to the varaibles w1, w2, w3 and w (bias)
w1 = np.random.uniform(-3.0, 3.0) 
w2 = np.random.uniform(-3.0, 3.0)
w3 = np.random.uniform(-3.0, 3.0)
w = np.random.uniform(-3.0, 3.0)

learning_rate = 0.5
max_iterations = 10000
k1 = np.linspace(-4, 4, num = 100)
k2 = np.linspace(-4, 4, num = 100)
k1_grid, k2_grid = np.meshgrid(range(-4,4), range(-4,4))
planes = []

# creating a 3D plane to show the animation of final classification plane respectively
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
ax.scatter(x1[t == 1], x2[t == 1], x3[t == 1], c='blue', marker='o', label ='Class 1', cmap='bwr')
ax.scatter(x1[t == 0], x2[t == 0], x3[t == 0], c='red', marker='s', label ='Class 0', cmap='bwr')
ax.legend()

# function to calculate sigmoid
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

# loop iterating 10000 times to get the final classification plane
for _ in range(max_iterations):
    
    # loop for calculating the error between the sigmoid output and target output
    error = 0
    for i in range(len(x1)):
        u = (w1 * x1[i]) + (w2 * x2[i]) + (w3 * x3[i]) + w
        o = sigmoid(u)
        error += (o - t[i])**2
    
    # if total error is less than or equal to 0.01 than break the loop
    if error <= 0.01:
        break
    
    # initializing varaibles for gradient descent of weights
    dw1 = 0
    dw2 = 0
    dw3 = 0
    dw = 0
    
    # loop to find the derivation of the the input array w.r.t the weight
    for i in range(len(x1)):
        u = (w1 * x1[i]) + (w2 * x2[i]) + (w3 * x3[i]) + w
        o = sigmoid(u)
        dw1 += learning_rate * (x1[i] * (o - t[i]) * (o * (1 - o)))
        dw2 += learning_rate * (x2[i] * (o - t[i]) * (o * (1 - o)))
        dw3 += learning_rate * (x3[i] * (o - t[i]) * (o * (1 - o)))
        dw += learning_rate * ((o - t[i]) * (o * (1 - o)))

    # updating weights
    w1 -= (1/5) * dw1
    w2 -= (1/5) * dw2
    w3 -= (1/5) * dw3
    w -= (1/5) * dw
    
    if dw1 == 0 or dw2 == 0 or dw3 == 0 or dw == 0:
        break
    
    # updating the classification plane using this equation
    if w3 != 0:
        k3 = -((w1 * k1_grid + w2 * k2_grid + w) / w3)
        
    # ax.view_init(elev=10.)
    
    # plotting the updated plane for every iteration
    # the code might take time to animate the classification plane, to check the output directly, comment the 5 lines below
    plane = ax.plot_surface(k1_grid, k2_grid, k3, alpha=0.01, cmap=cm.coolwarm)
    planes.append(plane)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.001)

# plotting the last plane on the graph
ax.plot_surface(k1_grid, k2_grid, k3, alpha=0.2, color='red')
plt.show()
```
