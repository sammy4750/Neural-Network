```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt

# input array
x = np.array([[1, 1], [1, -1], [-1, 1], [-1, -1], [5, 5], [-5, 5], [5, -5], [-5, -5]])
# output array
t = np.array([[0], [0], [0], [0], [1], [1], [1], [1]])

learning_rate = 0.1
max_iterations = 10000
number_of_nodes = 10
# Lists to store all the calculated errors in every iteration
error_history_list = []


# Activation function - Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# starting with 10 nodes in the first layer and then decreasing number of nodes by 1
for i in range(number_of_nodes, 0, -1):
    
    error_history_list = []

    # initializing random weights and bias for both the layers
    first_layer_weight = np.random.uniform(-3, 3, size=(2,i))
    first_layer_bias = np.random.uniform(-3, 3, size=(1,i))
    second_layer_weight = np.random.uniform(-3, 3, size=(i,1))
    second_layer_bias = np.random.uniform(-3, 3, size=(1,1))

    # Training loop (iterating 10000 times)
    for _ in range(max_iterations):
        # calculating forward propagation: x-->v-->z(f(v))-->u-->o(f(u))
        v = np.dot(x, first_layer_weight) + first_layer_bias
        z = sigmoid(v)
        u = np.dot(z, second_layer_weight) + second_layer_bias
        o = sigmoid(u)

        # Calculating the error
        error = (o - t)**2
        error_history_list.append(error.mean())

        # Backtracking to get the gradient descent to update the weights
        dw2 = (o - t) * (o * (1 - o))
        dw1 = np.dot(dw2, np.transpose(second_layer_weight)) * (z * (1 - z))

        # Updating weights and bias
        second_layer_weight -= learning_rate * np.dot(np.transpose(z), dw2)
        second_layer_bias -= learning_rate * np.sum(dw2)
        first_layer_weight -= learning_rate * np.dot(np.transpose(x), dw1)
        first_layer_bias -= learning_rate * np.sum(dw1)
        
    print('\nResult for {} number of nodes:'.format(i))

    # Printing input, weights, and output values
    print("\nInput to the first layer:")
    print(x)
    print("\nFinal updated weight for the first layer:")
    print(first_layer_weight)
    print("\nFinal updated bias for the first layer:")
    print(first_layer_bias)
    print("\nFinal updated weight for the second layer::")
    print(second_layer_weight)
    print("\nFinal updated bias for the second layer:")
    print(second_layer_bias)
    print("\nFinal Output after training 10000 times:")
    print(o)

    # Plot the history of the error function
    plt.plot(range(max_iterations), error_history_list)
    plt.title('Error vs. Iteration')
    plt.xlabel('Number of Iterations')
    plt.ylabel('Error')
    plt.show()
    print('-------------------------------------------------------------')
```
